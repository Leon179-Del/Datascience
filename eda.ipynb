```python
import pandas

data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
# use the Describe function
data.describe()
```

```python
# We have noticed that the column itemprice contains three empty slots
# We need to fill this empty slots with a value and in this case we shall fill with the mean of the column
data["itemprice"] = data["itemprice"].fillna(55)


# describe the data again
data.describe()
```

```python
# Get the number of entries in a given column
data.groupby("gender").size()
```

```python
# combining multiple columns
data.groupby(["city","payment"]).size()
```

```python
# import the modules
import seaborn as sns
import matplotlib.pyplot as pit
```

```python
# below is a count plot showing distribution of males to  female
sns.countplot(x="gender", data=data, palette="winter")
pit.title("Gender Distribution Against Total Count")
pit.ylabel("Total Count of Individuals")
```

```python
sns.countplot(x="product", data=data, palette="summer")
pit.xticks( rotation = 45)
```

```python
# below is a countplot that has two columns in it
sns.countplot(x="gender", hue="city", palette="Reds", data=data)
```

```python
# below is  a pieplot
data["payment"].value_counts().plot.pie(autopct='%1.1f%%')
```
